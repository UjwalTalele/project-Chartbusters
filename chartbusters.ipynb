```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("Data_Train.csv")
```

```python
train.head()
```

```python
train['day'] = pd.to_datetime(train['Timestamp']).dt.day
train['hour'] = pd.to_datetime(train['Timestamp']).dt.hour
train['month'] = pd.to_datetime(train['Timestamp']).dt.month
train['year'] = pd.to_datetime(train['Timestamp']).dt.year
train['minute'] = pd.to_datetime(train['Timestamp']).dt.minute
```

```python
train.head()
```

```python
missing = [col for col in train.columns if train[col].isnull().any()]
missing
```

```python
droppped = ['Country','Song_Name','Timestamp']
```

```python
train.drop(['Country','Song_Name','Timestamp'], axis =1)
```

```python
e_cols = ['Name','Genre']
```

```python
nunique = list(map(lambda col:train[col].nunique(),e_cols))
d = dict(zip(e_cols,nunique))
```

```python
print(sorted(d.items(),key = lambda x:x[1]))
```

```python
columns = list(train.columns)
columns.remove('Views')
```

```python
X =  train[columns]
X.head()
y = train['Views']
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
X['Name_enc'] = le.fit_transform(X['Name'])
X['Genre_enc'] = le.fit_transform(X['Genre'])
```

```python
X.drop(['Name','Genre','Country','Song_Name','Timestamp'],inplace = True,axis = 1)
```

```python
X.head()
```

```python

from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val = train_test_split(X,y,random_state = 1, test_size =0.2)
```

```python
def makeNumeric(data,col):
    X[col] = X[col].str.replace(',','')
    likes = []
    for d in X[col]:
        if d.endswith('K'):
            likes.append(float(d.replace('K',''))*1000)
        elif d.endswith('M'):
            likes.append(float(d.replace('M',''))*1000000)
        else:
            likes.append(float(d))
    return likes
```

```python
X['Likes'] = makeNumeric(X,'Likes')
X['Popularity'] = makeNumeric(X,'Popularity')
```

```python
x_train.head(10)
```

# training models

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

RFR_model = RandomForestRegressor(random_state = 1)
RFR_model.fit(x_train,y_train)
```

```python
y_pred_rf = RFR_model.predict(x_val)
```

```python
print(mean_squared_error(y_val,y_pred_rf))
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(y_val,y_pred_rf))
```

# Regression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred_reg  = reg.predict(x_val)
```

```python
print(mean_squared_error(y_pred_reg,y_val))
```

# XGboost

```python
import xgboost as xgb
xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10)
xg_reg.fit(x_train,y_train)
```

```python
y_pred = xg_reg.predict(x_val)
```

```python
print(mean_squared_error(y_pred,y_val))
```
